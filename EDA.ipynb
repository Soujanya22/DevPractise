```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
sns.set_context('talk')
sns.set_theme(style="darkgrid")


cpi_df = pd.read_csv("https://storage.googleapis.com/ml_bits/data.csv")

```

```python
cpi_df.columns
```

```python
cpi_df.describe()
```

```python
plt.figure(figsize=(40, 40))
matrix = np.triu(cpi_df.corr())
sns.heatmap(cpi_df.corr(), mask=matrix,annot=True)
```

```python
cpi_df.plot(kind='scatter', x='Year', y='General index')
```

```python
cpi_df.columns.unique()
```

```python
g = sns.relplot(x="Year", y="General index", kind="line", data=cpi_df)
g.fig.autofmt_xdate()
```
